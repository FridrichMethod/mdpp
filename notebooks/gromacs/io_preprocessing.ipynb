# `mdpp` Example: IO and Preprocessing

This notebook demonstrates the shared data-loading and preprocessing layer used by all other analysis notebooks:

1. `load_trajectory`
2. `select_atom_indices`
3. `align_trajectory`
4. `load_trajectories` for multi-run workflows
            


```python
from __future__ import annotations

from pathlib import Path

import matplotlib.pyplot as plt
from mplplots.utils import auto_ticks

from mdpp.analysis.trajectory import (
    align_trajectory,
    load_trajectories,
    load_trajectory,
    residue_ids_from_indices,
    select_atom_indices,
)

plt.style.use("mplplots.styles.GraphPadPrism")
```

## Input Files

Set paths to your trajectory and topology files before running analysis cells.
            


```python
TOPOLOGY_PATH = Path("/path/to/topology.pdb")
TRAJECTORY_PATH = Path("/path/to/trajectory.xtc")
STRIDE = 5
ATOM_SELECTION = "protein"

if not TOPOLOGY_PATH.exists() or not TRAJECTORY_PATH.exists():
    raise FileNotFoundError(
        "Update TOPOLOGY_PATH and TRAJECTORY_PATH before running analysis cells."
    )
```

## Load and Inspect a Trajectory


```python
traj = load_trajectory(
    trajectory_path=TRAJECTORY_PATH,
    topology_path=TOPOLOGY_PATH,
    stride=STRIDE,
    atom_selection=ATOM_SELECTION,
)

print(f"Frames: {traj.n_frames}")
print(f"Atoms: {traj.n_atoms}")
print(f"Time range (ps): {traj.time[0]:.3f} -> {traj.time[-1]:.3f}")
```

## Atom Selection and Residue Mapping


```python
ca_indices = select_atom_indices(traj.topology, "name CA")
ca_residue_ids = residue_ids_from_indices(traj.topology, ca_indices)

print(f"CA atoms selected: {ca_indices.size}")
print("First 10 residue IDs:", ca_residue_ids[:10])
```

## Align Trajectory to a Reference Frame


```python
aligned = align_trajectory(
    traj,
    atom_selection="name CA",
    reference_frame=0,
)

print(f"Aligned frames: {aligned.n_frames}")
print(f"Aligned atoms: {aligned.n_atoms}")
```
